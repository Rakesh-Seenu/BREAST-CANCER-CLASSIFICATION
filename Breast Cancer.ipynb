# MODEL TRAINING

```python

# Let's drop the target label coloumns
X = df_cancer.drop(['target'],axis=1)
```

```python
X.head()
```

```python
y = df_cancer['target']
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=5)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix
```

```python
svc_model = SVC()
```

```python
svc_model.fit(X_train, y_train)
```

# EVALUATING THE MODEL

```python
y_predict = svc_model.predict(X_test)
cm = confusion_matrix(y_test, y_predict)
```

```python
sns.heatmap(cm, annot=True)
```

```python
print(classification_report(y_test, y_predict))
```

# IMPROVING THE MODEL

```python
min_train = X_train.min()
min_train
```

```python
range_train = (X_train - min_train).max()
range_train
```

```python
X_train_scaled = (X_train - min_train)/range_train #normalization or feature scaling
```

```python
X_train_scaled
```

```python
sns.scatterplot(x = X_train['mean area'], y = X_train['mean smoothness'], hue = y_train)
```

```python
sns.scatterplot(x = X_train_scaled['mean area'], y = X_train_scaled['mean smoothness'], hue = y_train)
```

```python
min_test = X_test.min()
range_test = (X_test - min_test).max()
X_test_scaled = (X_test - min_test)/range_test
```

```python
from sklearn.svm import SVC 
from sklearn.metrics import classification_report, confusion_matrix

svc_model = SVC()
svc_model.fit(X_train_scaled, y_train)
```

```python
y_predict = svc_model.predict(X_test_scaled)
cm = confusion_matrix(y_test, y_predict)

sns.heatmap(cm,annot=True,fmt="d")
```

```python
print(classification_report(y_test,y_predict))
```

# IMPROVING THE MODEL USING C & Gamma Parameters

```python
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']} #Opti of c & gamma parameters
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV ( SVC(),param_grid,refit=True,verbose=4)
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test_scaled)
```

```python
cm = confusion_matrix(y_test, grid_predictions)
```

```python
sns.heatmap(cm, annot=True)
```
